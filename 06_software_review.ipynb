```python
# Ejercicio 2

#Desarrolla un programa que encuentre y muestre todos los números perfectos en el
#rango del 1 al 1,000,000. Un número perfecto es un entero positivo que es igual a la

# suma de sus divisores propios positivos. Los divisores propios son todos los
# divisores del número, excluyendo al propio número.
# El programa debe imprimir cada número perfecto encontrado.


for num in range(1, 1000001):
    suma_divisores = 0
    for i in range(1, num):
        if num % i == 0:
            suma_divisores += i
    if suma_divisores == num:
        print("Número perfecto encontrado:",num)
```

```python
# Ejercicio 3

# En un curso básico de Geometría Euclidiana se quiere construir un software que
# facilite el cálculo de áreas de ciertas figuras geométricas y algunos otros elementos
# importantes de la geometría básica. Para esto, se le pide que construya un software
# que pueda acceder a esas diferentes opciones a través de un menú. Que en cada
# submenú se pida al usuario que introduzca lo que se necesita para hacer los cálculos
# correspondientes.
# Este software será utilizado por personas con poca experiencia en informática, por
# lo que se le pide que sea fácil de maneja y que los mensajes sean muy intuitivos.
# El menú es el siguiente:
#** MENU **
#Área de un triangulo
#Área de un círculo
#N-ésimo término de Fibonacci
#Los 100 primeros números primos
#Cálculo del número PI
#Salir
#>> Ingrese una opción?

def areaTriangulo(base, altura):
    base=float(input("ingrese la base del triangulo: "))
    altura=float(input("ingrese la altura del triangulo: "))
    return (base * altura) / 2

def areaCirculo(radio):
    radio=float(input("ingrese el radio del circulo: "))
    return 3.1416 * (radio ** 2)
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a
def primerosPrimos(n):
    primos=[]
    num=2 # El primer número primo
    while primos>n:
        esPrimo=True
        for i in range(2,num):
            if num % i == 0:
                esPrimo=False
        if esPrimo:
            primos.append(num)
        num+=1
    return primos

def calcularPILeibniz(interaciones):
    pi_aprox=0
    for k in range(interaciones):
        termio=((-1)**k)/(2*k+1)
        pi_aprox+=termio
    pi_aprox*=4
    return pi_aprox

def menu():
    Opcion=0
    while Opcion != 6:
        print("\n** MENU **")
        print("1. Área de un triangulo")
        print("2. Área de un círculo")
        print("3. N-ésimo término de Fibonacci")
        print("4. Los 100 primeros números primos")
        print("5. Cálculo del número PI")
        print("6. Salir")
        Opcion=int(input(">> Ingrese una opción: "))
        if Opcion==1:
            base=float(input("ingrese la base del triangulo: "))
            altura=float(input("ingrese la altura del triangulo: "))
            print("El área del triángulo es:", areaTriangulo(base, altura))
        elif Opcion==2:
            radio=float(input("ingrese el radio del circulo: "))
            print("El área del círculo es:", areaCirculo(radio))
        elif Opcion==3:
            n=int(input("ingrese el valor de n: "))
            print(f"El {n}-ésimo término de Fibonacci es:", fibonacci(n))
        elif Opcion==4:
            print("Los 100 primeros números primos son:", primerosPrimos(100))
        elif Opcion==5:
            interaciones=int(input("ingrese el numero de iteraciones para aproximar PI: "))
            print("La aproximación de PI es:", calcularPILeibniz(interaciones))
        elif Opcion==6:
            print("Saliendo del programa...")
        else:
            print("Opción no válida, por favor intente de nuevo.")
menu()


```
